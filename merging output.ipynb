```python
psychopy_results = psychopy[['img_dilemma1','img_dilemma2','mouse.clicked_name','mouse.time','mouse.x','mouse.y']]
```

```python
psychopy_results
```

```python
dilemmi_main = psychopy_results['img_dilemma1']
dilemmi_choice = psychopy_results['img_dilemma2']
```

```python
gaze_mark = pd.read_csv('./output_psychopy/final_test_dilemmi_MARKER-MAPPER__csv/gaze.csv')
```

```python
events = pd.read_csv('./output_psychopy/raw-data-export/2023-12-12_17-44-53-9b9a4652/events.csv')
```

```python
start = np.array(events['timestamp [ns]'].loc[(events['name']=='recording.begin')])[0]
end = np.array(events['timestamp [ns]'].loc[(events['name']=='recording.end')])[0]
```

```python
fixations_mark = pd.read_csv('./output_psychopy/final_test_dilemmi_MARKER-MAPPER__csv/fixations.csv')
```

```python
events = events.loc[(events['name']=='dilemma_main') | (events['name']=='dilemma_choice') | (events['name']=='red_dot') | (events['name']=='red_dot2') | (events['name']=='end')].reset_index()
```

```python
events
```

```python
dilemma_time = list()
```

```python
counter_main = 0
counter_choice = 0
for n,row in enumerate(range(events.shape[0])):
    event_type = events.at[n,'name']
    if (event_type == 'dilemma_main'):
        event_begin = events.at[n,'timestamp [ns]']
        event_end = events.at[n+1,'timestamp [ns]']
        dilemma_time.append([event_type,event_begin,event_end,dilemmi_main[counter_main]])
        counter_main+=1
    elif (event_type=='dilemma_choice'):
        event_begin = events.at[n,'timestamp [ns]']
        event_end = events.at[n+1,'timestamp [ns]']
        dilemma_time.append([event_type,event_begin,event_end,dilemmi_choice[counter_choice]])
        counter_choice+=1
        
```

```python
dilemma_time
```

```python
for n,row in enumerate(range(gaze_mark.shape[0])):
    for n2,row2 in enumerate(dilemma_time):
        if (gaze_mark.at[n,'timestamp [ns]'] >= row2[1]) and (gaze_mark.at[n,'timestamp [ns]'] < row2[2] ):
            gaze_mark.at[n,'event'] = row2[0]
            gaze_mark.at[n,'img'] = row2[-1]
```

```python
for n,row in enumerate(range(fixations_mark.shape[0])):
    for n2,row2 in enumerate(dilemma_time):
        if (fixations_mark.at[n,'start timestamp [ns]'] >= row2[1]) and (fixations_mark.at[n,'end timestamp [ns]'] < row2[2] ):
            fixations_mark.at[n,'event'] = row2[0]
            fixations_mark.at[n,'img'] = row2[-1]
```

```python
gaze_mark = gaze_mark.loc[(gaze_mark['gaze detected on surface'] == True) & ((gaze_mark['event'] == 'dilemma_main')|(gaze_mark['event'] == 'dilemma_choice'))]
```

```python
fixations_mark = fixations_mark.loc[(fixations_mark['fixation detected on surface'] == True) & ((fixations_mark['event'] == 'dilemma_main')|(fixations_mark['event'] == 'dilemma_choice'))]
```
